```python
# length, height, width
# 길이,   높이,   너비,

import numpy as np

# 무게 타깃 (정답파일용)
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
# 훈련세트와 테스트 세트로 나눔

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    perch_full, perch_weight, random_state=42)

print(train_input[0], train_target[0])
# 판다스에서 받은 데이터 [length, height, width] weight
#                         [19.6   5.14    3.04]   85.
```

```python
# 훈련용 42개의 농어와 3개의 특성
# 테스트 14개의 농어와 3개의 특성

print(train_input.shape, test_input.shape)
# shape (행 개수, 열 개수)
```

```python
# [위 데이터를 사용해서 새로운 특성을 만들기]
# 사이킷런 변환기 : 특성을 만들거나 전처리하는 클래스가 있다.
# 변환기(Transformer) : fit(), transform() 메서드가 있다.

# 특성변환기
from sklearn.preprocessing import PolynomialFeatures
# 전에는 제곱을 이용해서 특성을 변경하여 곡선을 만들었다.
# PolynomialFeatures 사이킷런에서 수학자들이 만든 공식을 사용해 봄

# 기본 원리: 입력 구조
# PolynomialFeatures는 (n_samples, n_features) 형태를 기대한다.
# 행 = 샘플 개수
# 열 = 특성 개수
poly = PolynomialFeatures()

# 어떤 다항 특성을 만들지 정의하는 단계
poly.fit([[2, 3]]) # 해석 : 샘플 1, 특성 2개 2,3을 갖는다.
# 2개의 특성을 가진 2와 3으로 이루어진 샘플 하나를 적용 테스트
# 실무에서는 poly.fit_transform([[2, 3]]) 이렇게 사용해서 한번에 실행한다.

print(poly.transform([[2, 3]]))
# fit() 메서드와 transform() 메서드를 차례대로 실행(훈련을 해야 변환이 됨)
```

```python
# [[1. 2. 3. 4. 6. 9.]]
# 1 (bias)
# 2     # fit에 첫번째 값
# 3     # fit에 두번째 값
# 4 = 2 ** 2
# 6 = 2 * 3
# 9 = 3 ** 2

# PolynomialFeatures는 기본적으로 각 특성을 제곱한 항을 추가하고
# 특성끼리 서로 곱한 항을 추가함
# 2와 3을 각기 제곱한 4와 9가 추가 되고, 2와 3을 곱한 6이 추가됨
# 무게 = a * 길이 + b * 높이 + c * 두께 + d + 1
# 선형 방정식의 절편을 항상 값이 1인 특성과 곱해지는 계수임 -> (길이, 높이, 두께, 1)
```

```python
poly = PolynomialFeatures(include_bias=False)
#                           1 삭제 (자동으로 절편 항은 무시하지만 알아두면 좋음)
poly.fit([[2, 3]])
print(poly.transform([[2, 3]])) # 특성에 1이 무조건 계산되지 않기 때문에 속도향상됨
```

```python
# include_bias=True (기본값) | 절편을 자동으로 추가한다.
# → 항상 맨 앞에 1을 추가
# → 수식의 상수항(b0)을 표현하기 위함

# include_bias=False
# → 상수항 1을 만들지 않음 | 진짜 이유는 절 중복 방지다.

# 실무에서는 이렇게 사용한다.
# PolynomialFeatures(include_bias=False)
# LinearRegression(fit_intercept=True)
```

```python
# [어떤 항들이 만들어졌는지 확인하는 도구]

poly.get_feature_names_out()
# poly.get_feature_names_out(['length', 'height'])
# ['length' 'height' 'length^2' 'length height' 'height^2']
```

```python
# [실제 훈련시작용 생성]
poly = PolynomialFeatures(include_bias=False)
# 사이킷런의 선형 모델은 자동으로 bias를 추가하므로
# bias 특성을 만들 필요 없으므로 include_bias를 False로 지정한다.

poly.fit(train_input) # → “구조 정의”
train_poly = poly.transform(train_input) # → 실제 적용 후 [train_poly] 변수에 저장
print(train_poly.shape) # 42개의 요소에 9개의 특성(3개의 특성을 공학으로 9개로 늘림)
# 무게, 높이, 두께를 다중 결합하여 특성을 만듬 (특성공학)
```

```python
# 만들어진 특성을 확인
poly.get_feature_names_out()
```

```python
poly.get_feature_names_out(['length', 'height', 'width'])
#[ 'x0',   'x1',  'x2', 'x0^2', 'x0 x1', 'x0 x2', 'x1^2', 'x1 x2', 'x2^2']
# length, height, width length^2
#                            length*height
#                                     length*width
#                                                height^2
#                                                      height*width
#                                                                  width^2
```

```python
# 테스트용 세트 변환
test_poly = poly.transform(test_input)
```

```python
# 다중 회귀 모델 훈련하기
from sklearn.linear_model import LinearRegression #(선형회귀 모델)

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target)) # 훈련 세트 점수

# R²(결정계수) : 0.9903
# → “훈련 데이터 변동의 약 99.03%를 설명한다”는 뜻이지
# → “99.03% 정확도”라고 부르는 건 용어상 부정확해.

# 여기서 훈련 점수만 높은 경우 과적합이 발생될 수 있다.
# 테스트 R²가 떨어지는 경우 과적합이 가능성이 큰것이다.
```

```python
# 테스트 세트 점수
print(lr.score(test_poly, test_target))

# 97.14% 농어의 길이만을 사용했을 때에 있던 과소적합 문제는 나타나지 않음

# 모델이 너무 단순해서 데이터를 충분히 설명하지 못함
# 길이 하나만 사용하면 물고기의 부피 정보를 충분히 반영하지 못함

# height, width 추가 / 제곱항, 교차항 추가
# → 물리적으로 부피(Length × Height × Width)와 더 가까워짐

# 훈련 점수 99%, 테스트 점수가 97%이면 약간의 일반화 오차는 있지만
# 과소적합은 아님, 과적합도 심하지 않음

# 길이만 사용했을 때 발생했던 과소적합 문제가,
# 다중 특성과 다항 특성 확장으로 인해 개선되었다.
```

```python
# StandardScaler가 하는 일
# StandardScaler 싸이킷런 데이터 전처리 스케일 조정(스케일러) -> 표준점수로 변환시켜줌
# 예전에는 공식으로 평균과 표준편차를 사용하라고 했는데 StandardScaler를 사용하면 끝

#[스케일링이 필요한 이]
# 특성 간의 스케일 차이가 엄청 커지기 때문에 이 상태에서 회귀를 하면 큰 값 특성이 계수 계산에 과도하게 영향을 미친다.
# 행렬 계산이 불안정해진다.
# 과적합 문제가 심화된다.
# Ridge/Lasso 효과 왜곡


from sklearn.preprocessing import StandardScaler

ss = StandardScaler() # 훈련세트로 학습한 변환기를 사용해 테스트 세트까지 변환필수
ss.fit(train_poly)

train_scaled = ss.transform(train_poly) # 훈련용 데이터 스케일링
test_scaled = ss.transform(test_poly) # 테스트용 데이터 스케일링

# 표준 점수로 변환한 train_scaled, test_scaled 준비완료
# 평균과 표준편차로 공식을 적용할 필요가 없다. (표준점수화 됨)
```

```python
# 선형 회귀 모델에 규제를 추가한 모델은 2개가 있다.

# 릿지(ridge) : 계수를 제곱한 값을 기준으로 규제 (선호)
# 계수가 커질수록 패널티가 증가
# 모든 계수를 조금씩 줄인다.
# 0으로 만들지 않는다.

# 특징 : 모델을 부드럽게 만든다.
# 특징 : 다항회귀에서 매우 효과적
# 특징 : 다중공선성 문제 완화
# 전체를 고르게 줄이자

# 라쏘(lasso) : 계수의 절대값을 기준으로 규제를 적용 -> 0으로 제공 가능
# 일부 계수를 정확히 0으로 만든다
# 자동 특성 선택 효과
# 쓸모 없는 건 제거
```

```python
# [릿지 회귀 : 모델 객체를 만들고 fit() -> score() ]
from sklearn.linear_model import Ridge

ridge = Ridge()  # 릿지 클래스를 객체로 생성하고 변수 연결
ridge.fit(train_scaled, train_target) # fit로 구조를 정의한다.
print(ridge.score(train_scaled, train_target))
# 테스트결과 좋은 성능입을 판단할 수 있다. (모델 객체 생성시 alpha 매개변수로 규제의 강도를 조절함)
```

```python
print(ridge.score(test_scaled, test_target))
```

```python
# 릿지와 라쏘 모델을 사용할 때 규제의 양을 임으로 조절할 수 있다.
# 모델 객체를 만들 때 alpha 매개 값을 주면 된다.

# alpha의 의미
#alpha ↑ → 규제 강함 → 계수 작아짐 → 모델 단순 → 과소적합 가능
#alpha ↓ → 규제 약함 → 일반 선형회귀와 비슷 → 과적합 가능

# 적절한 alpha 값을 찾는 방법은 그래프로 그려 봐야 한다.
import matplotlib.pyplot as plt

train_score = []    # 훈련 :규제의 알파 농도따라서 결과 %를 기록저장용
test_score = []     # 테스트 % 저장용
# alpha 값을 바꿀때 마다 score() 메서드의 결과 값을 저장할 리스트 생성
```

```python
# alpha 값을 10배씩 증가시키며 실험
# alpha = 규제 강도 (클수록 계수를 더 강하게 줄임)
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

# 각 alpha 값에 대해 모델을 학습하고 성능을 비교
for alpha in alpha_list:

    # 1️⃣ Ridge 모델 생성
    # alpha가 클수록 규제가 강해짐
    ridge = Ridge(alpha=alpha)

    # 2️⃣ 훈련 데이터로 모델 학습
    # train_scaled : 표준화된 다항 특성
    # train_target : 실제 무게
    ridge.fit(train_scaled, train_target)

    # 3️⃣ 훈련 세트 R² 점수 저장
    # 과적합 여부 확인용
    train_score.append(
        ridge.score(train_scaled, train_target)
    )

    # 4️⃣ 테스트 세트 R² 점수 저장
    # 일반화 성능 확인용
    test_score.append(
        ridge.score(test_scaled, test_target)
    )

# 5️⃣ 각 alpha에 따른 성능 출력
print(f"train_score {train_score}")
print(f"test_score {test_score}")
```

```python
# -----------------------------
# alpha 값에 따른 R² 변화 시각화
# -----------------------------

# alpha 값은 0.001 ~ 100 처럼 지수적으로 증가하기 때문에
# 그대로 그리면 간격이 한쪽에 몰려 보인다.

plt.plot(np.log10(alpha_list), train_score)  # 훈련 세트 R²
plt.plot(np.log10(alpha_list), test_score)   # 테스트 세트 R²

# 따라서 log10을 사용해 지수 단위를 선형 축으로 변환한다.
# np.log10()
# np.log10(0.001)  → -3
# np.log10(0.01)   → -2
# np.log10(0.1)    → -1
# np.log10(1)      → 0
# np.log10(10)     → 1
# np.log10(100)    → 2

plt.xlabel('log10(alpha)')  # 실제로는 log 스케일
plt.ylabel('R^2')           # 회귀 모델의 결정계수

plt.show()
```

```python
# ---------------------------------------
#  최적 alpha 값을 사용하여 Ridge 모델 생성
# ---------------------------------------

# alpha=0.1 → 앞에서 그래프를 통해 테스트 점수가 가장 좋았던 값
# alpha는 규제 강도이며, 클수록 계수를 더 강하게 줄임
ridge = Ridge(alpha=0.1)

# 표준화된 훈련 데이터로 모델 학습
# train_scaled : 다항 확장 + 스케일링 완료된 훈련 입력
# train_target : 실제 무게 값
ridge.fit(train_scaled, train_target)


# ---------------------------------------
#  모델 성능 평가 (R² 점수)
# ---------------------------------------

# 훈련 세트 점수
# 모델이 훈련 데이터를 얼마나 잘 설명하는지 확인
print(ridge.score(train_scaled, train_target))

# 테스트 세트 점수
# 모델이 새로운 데이터에 대해 얼마나 일반화되는지 확인
print(ridge.score(test_scaled, test_target))


# ---------------------------------------
#  새로운 데이터 예측
# ---------------------------------------

# 예측할 새로운 물고기 데이터
# 형식은 반드시 (샘플 수, 특성 수) → 2차원 리스트
new_data = [[30.0, 8.0, 5.0]]


# ① 다항 특성 변환
# 학습할 때 사용한 동일한 PolynomialFeatures 객체 사용
# (3개 특성 → 55개 특성으로 확장)
new_poly = poly.transform(new_data)


# ② 스케일링
# 훈련 데이터 기준 평균과 표준편차로 변환
# 반드시 같은 StandardScaler 객체 사용
new_scaled = ss.transform(new_poly)


# ③ 예측
# 최종적으로 Ridge 모델에 입력하여 무게 예측
ridge.predict(new_scaled)
```

```python
# ---------------------------------------
#  Lasso 모델 생성
# ---------------------------------------
from sklearn.linear_model import Lasso

# Lasso는 L1 규제를 사용하는 선형 모델
# 기본 alpha 값은 1.0
# alpha가 클수록 규제가 강해지고, 일부 계수를 0으로 만들 수 있음
lasso = Lasso()

# ---------------------------------------
#  모델 학습
# ---------------------------------------

# 표준화된 다항 특성(train_scaled)으로 학습
# 반드시 스케일링 후 사용해야 함 (L1 규제는 스케일에 민감)
lasso.fit(train_scaled, train_target)

# ---------------------------------------
#  훈련 세트 R² 점수 확인
# ---------------------------------------

# 모델이 훈련 데이터를 얼마나 설명하는지 확인
print(lasso.score(train_scaled, train_target))
```

```python
# 지금 기본값도 안정적이지만 알파값을 조절하면 더 좋은 값을 찾을 수 있겠다???
train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    # 라쏘 모델을 만듭니다
    lasso = Lasso(alpha=alpha, max_iter=10000) # 최대 반복횟수(정확도를 위함)
    # 라쏘 모델은 최적의 계수를 찾기위해 반복적인 계산을 수행함

    # 라쏘 모델을 훈련합니다
    lasso.fit(train_scaled, train_target)

    # 훈련 점수와 테스트 점수를 위에 만든 리스트에 저장합니다
    train_score.append(lasso.score(train_scaled, train_target))
    test_score.append(lasso.score(test_scaled, test_target))

# 라쏘 테스트시 반복계산을 하는데 지정한 반복횟수가 부족하면 경고가 발생 ConvergenceWarning
```

```python
# 위에서 만든 리스트를 차트로 보자
plt.plot(np.log10(alpha_list), train_score) # 파란색
plt.plot(np.log10(alpha_list), test_score) # 주황색
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
# 왼쪽은 과대 적합, 오른쪽은 정상 (제일 좋은 값은 1 => 10)
```

```python
# ---------------------------------------
# Lasso(라쏘) 모델: alpha=10 적용
# ---------------------------------------

# Lasso는 L1 규제(계수의 절댓값 합에 패널티)를 사용
# alpha 값이 클수록 규제가 강해져서
# -> 계수들이 더 작아지고
# -> 일부 계수는 정확히 0이 되어 "특성 선택" 효과가 발생할 수 있음
#
# 참고: "log10(alpha)=1"이면 alpha=10 이라는 뜻 (10^1 = 10)
lasso = Lasso(alpha=10)

# 표준화된 훈련 데이터(다항특성 + 스케일링 완료)로 학습
# Lasso/Ridge는 스케일에 민감하므로 스케일링된 데이터를 쓰는 게 정석
lasso.fit(train_scaled, train_target)

# ---------------------------------------
# 성능 평가 (R² 점수)
# ---------------------------------------
# score()는 회귀에서 기본적으로 R²(결정계수)를 반환
# 훈련 점수: 훈련 데이터를 얼마나 잘 설명하는지
print(lasso.score(train_scaled, train_target))  # 예: 0.9888 (≈ 98.88%)

# 테스트 점수: 새로운 데이터에서의 일반화 성능
print(lasso.score(test_scaled, test_target))    # 예: 0.9824 (≈ 98.24%)

# ---------------------------------------
# 새로운 물고기 데이터 예측
# ---------------------------------------
# 원본 특성 3개 입력 (length, height, width)
new_data = [[30.0, 8.0, 5.0]]

# 1) 다항 특성 변환: (1, 3) -> (1, 55)
# 학습 때 사용한 poly(PolynomialFeatures) 객체로 동일하게 변환해야 함
new_poly = poly.transform(new_data)

# 2) 스케일링: 훈련 데이터 기준(평균/표준편차)으로 표준화
# 학습 때 사용한 ss(StandardScaler) 객체로 동일하게 변환해야 함
new_scaled = ss.transform(new_poly)

# 3) 예측: Lasso 모델로 무게 예측
lasso.predict(new_scaled)  # 예: array([331.95843724])
```
