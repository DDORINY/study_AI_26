```python
import numpy as np

# 각 학급에 3명의 학생, 각 학생이 3과목의 시험점수를 가지고 있는 배열 생성
# 예시 : 3개의 학급, 각 학습에 3명의 학생, 각 학생 당 국거, 영어, 수학 점수
scores =np.random.randint(0,101,size=(3,4,3))

# 학급별, 학생별, 과목별 점수를 출력
print("학급별 학생들의 점수 배열(3차원 배열):")
print(scores)
# 학급, 학생, 과목별로 점수 출력

for class_idx in range(scores.shape[0]):
    print(f"\nClass: {class_idx+1}")
    for student_idx in range(scores.shape[1]):
        korean,english,math=scores[class_idx,student_idx]
        print(f"Student {student_idx+1}-Korean:{korean}-English:{english}-Math:{math}")
```

```python
student1_scores=np.array([85,90,78])
student2_scores=np.array([92,88,95])
print("학생 1의 점수:",student1_scores)
print("학생 2의 점수:",student2_scores)
total_scores =student1_scores+student2_scores
print("두 학생의 총점:",total_scores)
avg_scores=total_scores/2
print("두 학생의 평균 점수:",avg_scores)
student1_total=np.sum(student1_scores)
student2_total=np.sum(student2_scores)
student1_avg=np.mean(student1_scores)
student2_avg=np.mean(student2_scores)

print("학생 1의 총점:",student1_total)
print("학생 1의 평균 점수:",student1_avg)
print("학생 2의 총점:",student2_total)
print("학생 2의 평균 점수:",student2_avg)
```
