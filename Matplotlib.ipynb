```python
# 데이터 시각화 라이브러리, 맷플롯립(Matplotlib)
# 타이타닉 데이터를 활용하여
# 01. 맷플롯립 기본 개념 이해
# 02. 다양한 그래프 생성 방법 학습
# 03. 그래프 스타일 및 옵션 설정
# 04. 데이터 시각화를 통해 인사이트 도출

# Matplotlib은 파이썬에서 데이터를 시각화하기 위한 대표적인 라이브러리입니다.
# Matplotlib: Visualization with Python
```

```python
# 맷플롯립 사용하기
import matplotlib.pyplot as plt
```

```python
# 간단한 맷플롯립으로 그래프를 만들고 출력해본다.
# 첫 번째 리스트의 값은 X값, 두 번째 리스트의 값은 y로 적용
# 순서쌍 (x,y)으로 매칭된 값을 좌표평면 위에 그래프 시각화
plt.plot([1,2],[2,4])
plt.show()
```

```python
# plot()함수에 리스트 하나만 전달하여 그래프 그리기

# 하나의 숫자 리스트 입력
# 리스트의 값들이 y값들이라고 가정하고 x값 [0,1,2,3]을 자동으로 만들어냄
# plt.show() 함수는 그래프를 화면에 나타나도록 함

y = [2,3,4,5]
plt.plot(y)
plt.show()
```

```python
# 두 개의 숫자 리스트 입력
# 첫 번째 리스트의 값은 x값, 두 번째 리스트의 값은 y로 적용됨
# 순서쌍(x,y)으로 매칭된 값을 좌표평면 위에 그래프 시각화

x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
# 범례 지정하기
# 범례는 그래프에서 데이터 종류를 표시하는 텍스트이다.

plt.plot([1,2,3,4],[1,4,9,16],label = 'Square')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend()
plt.show()
```

```python
# 범례의 위치를 위쪽 가운데로 지정하고 싶다면

# linspace() 함수로 넘파이 배열 생성
import numpy as np
x = np.linspace(0,10,100)
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

# 그래프 그리기
plt.plot(x,y1,label = 'y =x')
plt.plot(x,y2,label = 'y =2x')
plt.plot(x,y3,label = 'y =0.5x')
plt.plot(x,y4,label = 'y =3x')

# 범례 설정(ncol을 2로 설정하여 한 줄에 2개 표시)
plt.legend(loc='upper center', ncol=2)

plt.title('Simple Liner Plot Example')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
# 축 범위 지정하기

plt.plot([1,2,3,4],[3,6,9,12])
plt.xlim([0,5])
plt.ylim([0,15])
plt.show()
```

```python
plt.plot([1,2,3,4,],[3,6,9,12])
plt.axis([0,5,0,15])
plt.show()
```

```python
# 다양한 옵션으로 하나씩 확인하며 그려보기

x = np.linspace(1,10,100)
y = np.exp(x)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Liner Scale')

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale('log')
plt.title('Logarithmic Scale (y-axis)')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.xscale('log')
plt.title('Log-Linear Scale (x-axis)')

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale('log')
plt.xscale('linear')
plt.title('Linear-Logarithmic Scale')

plt.tight_layout()
plt.show()
```

```python
plt.plot([1,2,3],[4,4,4], '-', label='solid')     # 실선
plt.plot([1,2,3],[3,3,3], '--', label='dashed')   # 점선(긴)
plt.plot([1,2,3],[2,2,2], ':', label='dotted')    # 점선(짧은)
plt.plot([1,2,3],[1,1,1], '-.', label='dash-dot') # 점-선

plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper center', ncol=4)
plt.show()
```

```python
plt.plot([1,2,3],[4,4,4], linestyle='dotted', label='dotted')
plt.plot([1,2,3],[3,3,3], linestyle='dashdot', label='dash-dot')
```

```python
# 마커(marker) 설정
# 선 그래프에서 각 데이터 포인트를 표시하는 기호입니다.

plt.plot([4,5,6], "b")     # 파란색
plt.plot([3,4,5], "ro")    # red + circle
plt.plot([2,3,4], marker="s")   # square
plt.plot([1,2,3], marker="D")   # diamond
plt.plot([0,1,2], marker="$A$")
plt.show()
```

```python
plt.plot([1,2,3,4],[2.0,3.3,6.3,10.5],color='violet')
plt.plot([1,2,3,4],[2.0,3.1,5.3,8.5],color='g')
plt.plot([1,2,3,4],[2.0,2.8,4.3,6.5], color= (0.1,0.2,0.3))
plt.plot([1,2,3,4],[2.0,2.5,3.3,4.5],color='#FF0000')
plt.show()
```

```python
# 2024년 팬톤 컬러 :peach Fuzz(#F5C8B0)
pantone_2024_color ='#F5C8B0'

# 데이터 생성
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.figure(figsize=(10, 6))
plt.plot(x,y, color = pantone_2024_color,linestyle='-', marker='o')
plt.title('Line Graph with 2024 Pantone color')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
# Matplotlib에서 그래프 제목(title)을 설정하는 방법
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.title('Graph Title',
          fontsize=16,
          color='blue',
          fontweight='bold',
          loc='center',
          pad=100,
          backgroundcolor='lightgray')

plt.show()
```

```python
# 슈퍼 타이틀 설정하기
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,2,1,2,1]

# 첫 번째 서브 그래프
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('Graph Sub Title 1', loc='left')

# 두 번째 서브 그래프
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('Graph Sub Title 2', loc='left')

# 슈퍼 타이틀 설정 (전체 figure 제목)
plt.suptitle('Graph Super Title', fontsize=16)

# 서브 그래프 간격 자동 조정
plt.tight_layout()

plt.show()
```

```python
# 눈금 설정하기
import matplotlib.pyplot as plt

x = [1,2,3]
years = ['2030','2040','2050'] # year는 x축에 표시할 연도
values = [300,100,700] #values는 y축에 표시할 값

plt.bar(x,values,color=['r','g','b'])
# x축 눈금에 '2030', '2040', '2050'를 표시
plt.xticks(x,years)
plt.show()
```
