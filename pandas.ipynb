```python
# 데이터 필터링하기
# 비교연산자 >로 필터링하기
netflix['release_year'] > 2015
```

```python
# 2015년 이후에 나온 콘텐츠 추출
more2015 = netflix[netflix['release_year'] > 2015]
more2015.head()
```

```python
# 부정연산자 ~로 필터링하기
less2015 = netflix[~(netflix['release_year'] > 2015)]
less2015.head()
```

```python
# 논리 연산자 &나 | 로 필터링 하기
# 2015년 이후에 나온 콘텐츠이면서(&) 쇼의 장리가 TV show인 데이터 추출
(netflix['release_year'] > 2015) &(netflix['type'] == 'TV Show')
```

```python
# 이것을 활용하면 원하는 값을 추출할 수 있습니다.
more2015_tv = netflix[(netflix['release_year'] > 2015) & (netflix['type'] == 'TV Show')]
more2015_tv.head()
```

```python
# 이번에는 | 연산자를 사용해서 2015년 이후에 나온 콘텐츠이거나  쇼의 장르가 TV show인 데이터 추출
more2015_or_tv = netflix[(netflix['release_year'])|(netflix['type'] == 'TV Show')]
more2015_or_tv.head()
```

```python
# 결측치 처리하기
# 판다스의 결측치는 NaN,None이 있습니다.

# isna().sum()를 이용한 결측치 확인하기
# 결츨치 확인 (1)-.isna().sum()
netflix.isna().sum()
```

```python
# info()함수를 이요한 비 결측치 확인하기
# 결측치 확인 (2)-.info()
netflix.info()
```

```python
# for문으로 결측치 비율 확인
# 결측치 확인 (3) - for문
for i in netflix.columns :
    missingValueRate = netflix[i].isna().sum() / len(netflix)*100
    if missingValueRate > 0 :
        print("{} null rate:{}%".format(i, round(missingValueRate, 2)))
```

```python
# 결측치 처리하기
# fillna()함수로 결측히를 다른 값으로 대체하기

# 결측치 처리 (1) - fillna()
# 결측치 비율 : country(25.0%) : 8개 중에 2 개 결측치
netflix['country']= netflix['country'].fillna("No data")
netflix
```

```python
# replace()함수로 결측치를 다른 값으로 대체하기
# 결측치 처리 (2) - replace()
# 결측치 비율 : director(37.5%) : 8개 중에 3개 결측치

netflix['director']=netflix['director'].replace(np.nan,'NO Data')
netflix
```

```python
# dropna()로 결측치를 가진 열 제거하기
# 결측치 처리 (3) -dropna()
# 결측치 비율 : listed_in(75.0%):8개 중에 6개의 결측치

netflix.dropna(axis=1,inplace=True)
netflix
```

```python
netflix = pd.read_csv('netflix_cleaned.csv')
print('mean :', netflix['duration'].mean())
print('median :', netflix['duration'].median())
print('sum :',netflix['duration'].sum())
print('min :', netflix['duration'].min())
print('max :',netflix['duration'].max())
```

```python
# 표준편차 :std()함수
# 분산 : var()함수

# duration 열의 표준 편차, 분산구하기
print('std : ', netflix['duration'].std())
print('var : ', netflix['duration'].nunique())
```

```python
# 비어 있지 않은 값 세기 :count()함수
# 열의 빈도수: value_counts()함수
# type 열의 비어 잇지 않은 값의 개수와 빈도수 구하기
print('count : ', netflix['type'].count())
print('value_counts : ', netflix['type'].value_counts())
```
